```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sun_data = pd.read_csv('../.run/sun_position.txt', sep=' ', header=None, names=['x', 'y'])
mercury_data = pd.read_csv('../.run/mercury_position.txt', sep=' ', header=None, names=['x', 'y'])
venus_data = pd.read_csv('../.run/venus_position.txt', sep=' ', header=None, names=['x', 'y'])
earth_data = pd.read_csv('../.run/earth_position.txt', sep=' ', header=None, names=['x', 'y'])
moon_data = pd.read_csv('../.run/moon_position.txt', sep=' ', header=None, names=['x', 'y'])
mars_data = pd.read_csv('../.run/mars_position.txt', sep=' ', header=None, names=['x', 'y'])
jupiter_data = pd.read_csv('../.run/jupiter_position.txt', sep=' ', header=None, names=['x', 'y'])
saturn_data = pd.read_csv('../.run/saturn_position.txt', sep=' ', header=None, names=['x', 'y'])
uranus_data = pd.read_csv('../.run/uranus_position.txt', sep=' ', header=None, names=['x', 'y'])
neptune_data = pd.read_csv('../.run/neptune_position.txt', sep=' ', header=None, names=['x', 'y'])
```

```python
plt.grid(True, ls='--')
plt.axis('equal')

plt.plot(sun_data['x'], sun_data['y'], label='Sun', marker='o', markersize=8)
plt.plot(mercury_data['x'], mercury_data['y'], label='Mercury')
plt.plot(venus_data['x'], venus_data['y'], label='Venus')
plt.plot(earth_data['x'], earth_data['y'], label='Earth')
plt.plot(mars_data['x'], mars_data['y'], label='Mars')
plt.plot(jupiter_data['x'], jupiter_data['y'], label='Jupiter')
plt.plot(saturn_data['x'], saturn_data['y'], label='Saturn')
plt.plot(uranus_data['x'], uranus_data['y'], label='Uranus')
plt.plot(neptune_data['x'], neptune_data['y'], label='Neptune')

plt.legend()

# plt.xlim((-1.5e11, 1.5e11))
# plt.ylim((-1.5e11, 1.5e11))
```

```python
sun_data = pd.read_csv('../../.run/sun_position.txt', sep=' ', header=None, names=['x', 'y'])
earth_data = pd.read_csv('../../.run/earth_position.txt', sep=' ', header=None, names=['x', 'y'])
moon_data = pd.read_csv('../../.run/moon_position.txt', sep=' ', header=None, names=['x', 'y'])
```

```python
# create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# supbplot 1: Earth and Moon orbits around Sun
axs[0].grid(True, ls='--')
axs[0].axis('equal')
axs[0].set_title('Earth and Moon Orbits Around Sun')
axs[0].plot(sun_data['x'], sun_data['y'], label='Sun', marker='o', markersize=8)
axs[0].plot(earth_data['x'], earth_data['y'], label='Earth')
axs[0].plot(moon_data['x'], moon_data['y'], label='Moon Absolute')
axs[0].legend()

# subplot 2: Moon orbit around Earth
axs[1].grid(True, ls='--')
axs[1].axis('equal')
axs[1].set_title('Moon Orbit Around Earth')
axs[1].plot(0, 0, label='Earth', marker='o', markersize=8)
axs[1].plot(moon_data['x'] - earth_data['x'], moon_data['y'] - earth_data['y'], label='Moon')
axs[1].legend()

axs[0].set_xlim((-1.5e11, 1.5e11))
axs[0].set_ylim((-1.5e11, 1.5e11))
axs[1].set_xlim((-5e8, 5e8))
axs[1].set_ylim((-5e8, 5e8))
# plt.xlim((-3e11, 3e11))
# plt.ylim((-3e11, 3e11))
# plt.xlim((-1.5e11, 1.5e11))
# plt.ylim((-1.5e11, 1.5e11))

# plt.xlim((-1e9, 1e9))
# plt.ylim((-1e9, 1e9))
```
